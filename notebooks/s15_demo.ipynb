```python
name = "arjun masciarelli"

freq = {}
for c in name:
    if c not in freq:  # first time seeing this letter
        freq[c] = 1
    else:
        freq[c] += 1
```

```python
name = "arjun masciarelli"

freq = {}
for c in name:
    freq[c] = freq.get(c, 0) + 1
```
