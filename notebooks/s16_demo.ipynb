## Tuples

```python
# Tuples are similar to lists, but they are immutable (cannot be changed after creation).

t = (3, 50, 54)
```

```python
t[0]
```

```python
t[-1]
```

```python
t[1:]
```

```python
len(t)
```

```python
sorted(t)
```

```python
t
```

```python
for num in t:
    print(num * 2) 
```

```python
t[1] = 55
```

```python
# create a dict to show the date - holiday name mapping
holidays = {
    (1,1): 'New Year', 
    (12, 25): 'Christmas',
    (3, 14): 'Pi Day'
}
```

```python
holidays[(1,1)]
```

```python
holidays[(3,14)]
```

## Mutable types cannot be used as dict keys.

```python
# how to get the date for pi day?

for date, name in holidays.items():
    if name == 'Pi Day':
        print(date)
```
